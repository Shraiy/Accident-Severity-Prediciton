```python
# Function to process Zipcode
def process_zipcode(zipcode):
    if isinstance(zipcode, str) and ' - ' in zipcode:  # Check if it's a range
        try:
            zip_start = int(zipcode.split(' - ')[0])  # Extract the first value before '-'
            return zip_start  # Return the first zip code
        except ValueError:
            return np.nan  # Handle errors
    try:
        return int(zipcode)  # Convert single zip code to integer
    except ValueError:
        return np.nan  # Handle any unexpected values

# Apply function to Zipcode column
data_accident_severity['Zipcode'] = data_accident_severity['Zipcode'].apply(process_zipcode)

# Display updated DataFrame
print(data_accident_severity)
```

```python
import re
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
from sklearn.feature_extraction.text import TfidfVectorizer
from scipy.sparse import hstack

# Initialize Stemmer
stemmer = PorterStemmer()

# Function for Text Preprocessing
def preprocess_text(text):
    if pd.isna(text):  # Handle missing values
        return ""
    text = text.lower()  # Convert to lowercase
    text = re.sub(r'\W+', ' ', text)  # Remove punctuation and special characters
    tokens = word_tokenize(text)  # Tokenization
    stemmed_tokens = [stemmer.stem(word) for word in tokens]  # Stemming
    return ' '.join(stemmed_tokens)  # Join words back

# Apply Text Preprocessing to Description Column
data_accident_severity['Description'] = data_accident_severity['Description'].apply(preprocess_text)

# Convert "Description" into TF-IDF Features
vectorizer = TfidfVectorizer(max_features=500)  # Limit to 500 most important words
X_tfidf = vectorizer.fit_transform(data_accident_severity['Description'])  # Sparse Matrix

# Drop Original "Description" Column
data_accident_severity.drop(columns=['Description'], inplace=True)

# Convert Other Features to Sparse Format (to save memory)
X_other = data_accident_severity.drop(columns=['Severity'])  # Drop target column
X_other_sparse = X_other.astype(np.float32)  # Convert to float32

# Combine TF-IDF Features with Other Features
X_final = hstack([X_other_sparse, X_tfidf])  # Efficiently merge sparse matrices
y = data_accident_severity['Severity']  # Target variable
```

```python
# Check for null values in X_final and y
print("Null values in X_final:\n", np.isnan(X_final.data).sum())
print("\nNull values in y:\n", y.isnull().sum())
```

```python
from sklearn.model_selection import train_test_split

# Split Data
X_train, X_test, y_train, y_test = train_test_split(X_final, y, test_size=0.2, random_state=42, stratify=y)
```

```python
print(np.isnan(X_train.toarray()).sum())  # Count total NaNs
```

```python
from sklearn.feature_selection import VarianceThreshold

# Remove features with low variance (threshold can be tuned)
selector = VarianceThreshold(threshold=0.01)
X_train = selector.fit_transform(X_train)
X_test = selector.transform(X_test)
```

```python
from imblearn.over_sampling import SMOTE

# Apply SMOTE to the training set
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

print(X_train_resampled.shape)
print(y_train_resampled.shape)
print(y_train_resampled.value_counts())
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score

# Initialize and train the Logistic Regression model
logreg_model = LogisticRegression(max_iter=1000)  # Increased max_iter
logreg_model.fit(X_train_resampled, y_train_resampled)

# Make predictions on the test set
y_pred = logreg_model.predict(X_test)

# Evaluate the model
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier

# Initialize and train the Random Forest Classifier
rf_classifier = RandomForestClassifier(random_state=42)  # You can adjust hyperparameters
rf_classifier.fit(X_train_resampled, y_train_resampled)

# Make predictions
y_pred_rf = rf_classifier.predict(X_test)

# Evaluate the model
print("Random Forest Accuracy:", accuracy_score(y_test, y_pred_rf))
print(classification_report(y_test, y_pred_rf))
```

```python
import xgboost as xgb

# Initialize and train the XGBoost model
xgb_model = xgb.XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='mlogloss') # You can adjust hyperparameters
xgb_model.fit(X_train_resampled, y_train_resampled)

# Make predictions
y_pred_xgb = xgb_model.predict(X_test)

# Evaluate the model
print("XGBoost Accuracy:", accuracy_score(y_test, y_pred_xgb))
print(classification_report(y_test, y_pred_xgb))
```

```python
from sklearn.ensemble import ExtraTreesClassifier

# Initialize and train the Extra Trees Classifier
et_classifier = ExtraTreesClassifier(random_state=42)  # You can adjust hyperparameters
et_classifier.fit(X_train_resampled, y_train_resampled)

# Make predictions
y_pred_et = et_classifier.predict(X_test)

# Evaluate the model
print("Extra Trees Accuracy:", accuracy_score(y_test, y_pred_et))
print(classification_report(y_test, y_pred_et))
```
